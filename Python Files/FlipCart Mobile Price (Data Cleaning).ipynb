```python
#Import the libraries
import pandas as pd
import numpy as np
import math
import datetime
import re

```

```python
#Data Reading
df = pd.read_csv('/content/mobile_prices_2023.csv')
df.head()
```

### In this project we should start with data cleaning like:
### First let's check for null values

```python
df.isnull().sum()
```

### If we look at the data some values from the columns are mixed with the other columns. which is why we need to fix it rather than deleting the data.

```python
df.sample(10)
```

### Let's grab the phone company name first from the column phone name

```python
df['Company Name'] = df['Phone Name'].str.split(" ").str[0]
```

```python
df['Company Name'].unique()
```

### As we can see that some values like Realme and realme comes again and again and mostly phone names are not the name just phone name so we need to change this.

```python
#Merge the realme and Realme
df['Company Name'] = df['Company Name'].str.replace('Realme','realme')
df['Company Name'] = df['Company Name'].str.replace('Apple','APPLE')
df['Company Name'] = df['Company Name'].str.replace('Redmi','REDMI')
df['Company Name'] = df['Company Name'].str.replace('Xiaomi','Mi')
df['Company Name'] = df['Company Name'].str.replace('Motorola','MOTOROLA')
df['Company Name'] = df['Company Name'].str.replace('Moto','MOTOROLA')
```

```python
df['Company Name'].unique()
```

### Now remove the other values like '10A'

```python
df = df[(df['Company Name'] != '10A') & (df['Company Name'] != 'a')  & (df['Company Name'] != '�9A') & (df['Company Name'] != 'A10E')]
```

```python
df['Company Name'].unique()
```

```python
df.sample(5)
```

```python
#check point
df1 = df.copy()
```

### In number of Ratings column the values are in object format that should be in int.

```python
df1['Number of Ratings'] = df1['Number of Ratings'].replace({',':''}, regex = True)
df1.head(2)
```

```python
df1['Number of Ratings'] = df1['Number of Ratings'].astype(int)
df1['Number of Ratings'].dtype
```

### Now let's extract the Ram size from the RAM column

```python
df1['Ram_size'] = df1['RAM'].str.split(' ').str[0]
df1.head(2)
```

```python
df1['Ram_size'].unique()
```

```python
df1 = df1[(df1['Ram_size'] != 'NA') & (df1['Ram_size'] != 'Expandable') & (df1['Ram_size'] != 'cm') & (df1['Ram_size'] != '768')]
```

```python
df1['Ram_size'].unique()
```

```python
#check point
df2 = df1.copy()
```

### Now let's extract the Storage size from Rom column

```python
df2['Storage_size'] = df2['ROM/Storage'].str.split(' ').str[0]
```

```python
df2['Storage_size'].isnull().sum()
```

### Now extrating the values for Battery column

```python
df2['Battery_size'] =df2['Battery'].str.split(" ").str[0]
df2['Battery_size'].unique()
```

### Remove the wrong values

```python
df2 = df2[(df2['Battery_size'] != 'A15') & (df2['Battery_size'] != 'A13') & (df2['Battery_size'] != 'A14') & (df2['Battery_size'] != 'A16') & (df2['Battery_size'] != 'A12') &
    (df2['Battery_size'] != 'A9') & (df2['Battery_size'] != 'Apple') & (df2['Battery_size'] != '1') & (df2['Battery_size'] != '0') & (df2['Battery_size'] != 'nan') &
    (df2['Battery_size'] != 'MediaTek') & (df2['Battery_size'] != 'Unisoc') & (df2['Battery_size'] != '2') & (df2['Battery_size'] != 'Brand')]
```

```python
df2['Battery_size'].unique()
```

```python
df2['Price in INR'] = df2['Price in INR'].replace({'₹':''}, regex = True)
df2['Price in INR'] = df2['Price in INR'].replace({',':''}, regex = True)
```

```python
df2.head()
```

```python
#check point
df3 = df2.copy()
```

```python
df2['Company Name'].unique()
```

### Now Let's change the datatypes of columns

```python
df3.info()
```

```python
df3['Price in INR'] = df3['Price in INR'].astype(int)
df3['Ram_size'] = df3['Ram_size'].astype(float)
```

```python
#Date column
df3['Date of Scraping'] = pd.to_datetime(df3['Date of Scraping'])
```

### Remove unwanted columns

```python
cols_list = ['RAM','ROM/Storage','Battery']
df3.drop(cols_list,axis=1,inplace=True)
df3.head()
```
